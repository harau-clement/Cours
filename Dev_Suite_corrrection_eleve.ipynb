## Exercice 3 : Mur triangulaire

Sujet1-Ex3.png

Le programme suivant de créer une suite arithmétique, d'afficher ses termes
et de calculer la somme de ceux-ci.

Adapter l'instruction `for n in range(1,7):` pour pouvoir répondre aux questions de l'exercice 

```python
premier_terme = int(input("Quel est le premier terme ?"))
raison= int(input("Quelle est la raison ?"))
suite = [0] # Création d'une liste qui va stocker tout les termes de la suite

for n in range(1,5):
    suite.append(premier_terme+(n-1)*raison) # enregistrement du terme dans la liste
    print("Terme",n,":",suite[n]) # affichage du terme 

print("La suite est ",suite) 

print("La somme des termes de cette suite vaut",sum(suite))
```
